```python
from tensorflow.keras.models import load_model
import PIL
import tensorflow as tf
import numpy as np
```

```python
faceNetModel = load_model('/content/drive/MyDrive/Colab Notebooks/myFaceRec/facenet_keras.h5')
faceNetModel.load_weights('/content/drive/MyDrive/Colab Notebooks/myFaceRec/facenet_keras_weights.h5')
```

```python
print("input:",faceNetModel.inputs)
print("output:",faceNetModel.outputs)
```

```python
def img_to_encodings(model, img_path):
  img = tf.keras.preprocessing.image.load_img(img_path, target_size=(160,160))
  img = np.around(np.array(img) / 255, decimals = 12)
  img = np.expand_dims(img, axis=0)
  encoding = model.predict(img)
  return encoding/np.linalg.norm(encoding, ord=2)
```

```python
pavel_img_path = '/content/drive/MyDrive/Colab Notebooks/myFaceRec/images/Pavel1m.jpg'
pavel_encodings = img_to_encodings(faceNetModel,pavel_img_path)
face_database = {}
face_database['Pavel'] = pavel_encodings
```

```python
def recognize(model, img_path, database):
  current_img_encoding = img_to_encodings(model, img_path)
  closest_dist = 100
  closest_identity = ''
  recognition_trashold = 0.7
  for identity, encoding in database.items():
    current_distance = np.linalg.norm(encoding - current_img_encoding)
    if current_distance < closest_dist:
      closest_dist = current_distance
      closest_identity = identity
      print(closest_dist)

  if closest_dist < recognition_trashold:
    print("Recognized person is:", closest_identity)
  else:
    print("Not recognized")
```

```python
recognize(faceNetModel, '/content/drive/MyDrive/Colab Notebooks/myFaceRec/images/Pavel3m.jpg', face_database)
```
